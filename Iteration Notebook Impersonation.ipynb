```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob, os
```

```python
# AWID CLS R Trn - Loading the dataset's folder
path = r'/mnt/data/dino/dataset/'


csvs = "AWID-CLS-R-Trn/1"
csv_paths = os.path.join(path, csvs)
csv_glob = glob.glob(csv_paths)

header_path = os.path.join(path, 'header.csv')
header = pd.read_csv(header_path, header=None)

awid_arr = []
for csv in csv_glob:
    awid_arr.append(
        pd.read_csv(csv, header=None, low_memory=False,
                    names=header.iloc[0]
                   )        
     )
awd = pd.concat(awid_arr)
```

# Loading Subset of Reduced Dataset

```python
awid = awd[ (awd['class'] == 'impersonation') | (awd['class'] == 'normal') ]
```

```python
awid["class"].value_counts()
```

```python
awid.info()
```

# Subset of Dataset

```python
#attack = awid[awid['class'] == 'impersonation'][:1000]
#attack.shape

attack = awid[awid['class'] == 'impersonation'].sample(1000)
attack.shape
```

```python
#benign = awid[awid['class'] == 'normal'][:10000]
#benign.shape
benign = awid[awid['class'] == 'normal'].sample(10000)
benign.shape
```

```python
awid_df = pd.concat([attack, benign])
awid_df.shape
```

```python
awid_df
```

```python
awid_df["class"].value_counts()
```

# Replace ? with NaN and Drop NaN values

```python
#awid.replace({'?':np.nan}).dropna(axis=1,inplace=True)
#awid.super()replace('?',np.nan,inplace=True)
```

```python
awid_df.head()
```

```python
awid_df=awid_df.replace('?',np.nan)
```

```python
awid_df.head()
```

```python
awid_df=awid_df.dropna(axis=1)
```

```python
awid_df.info()
```

```python
awid_df.head()
```

```python
awid_df.info()
```

```python
awid_df.loc[ awid_df['class']=='impersonation','class' ]=1
awid_df.loc[ awid_df['class']=='normal','class' ]=0
awid_df['class'] = awid_df['class'].astype(int)
```

```python
awid_df['class'].value_counts()
```

```python
awid_df['radiotap.present.reserved'].value_counts()
```

```python
#awid_df.drop(["radiotap.present.reserved", "wlan.fc.type_subtype", "wlan.fc.ds"], axis=1, inplace=True)

for f in ["radiotap.present.reserved", "wlan.fc.type_subtype", "wlan.fc.ds"]:
    awid_df[f] = awid_df[f].apply(int, base=16)
```

```python
awid_df.info()
```

```python
awid_df['class'].value_counts()
```

# Sep features and target columns


```python
train_attack_no = 992
train_benign_no = 1000

awid_train_attack = awid_df[awid_df['class'] == 1][:train_attack_no]
awid_train_benign = awid_df[awid_df['class'] == 0][:train_benign_no]

awid_train = pd.concat([awid_train_attack, awid_train_benign])


awid_test_attack = awid_df[awid_df['class'] == 1][train_attack_no+1:]
awid_test_benign = awid_df[awid_df['class'] == 0][train_benign_no+1:]

awid_test = pd.concat([awid_test_attack, awid_test_benign])
```

# Train and Test Split

```python
X_train = awid_train.drop( ["class"], axis=1)
X_test = awid_test.drop( ["class"], axis=1)
y_train = awid_train['class']
y_test = awid_test['class']
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#clf = KNeighborsClassifier()
#clf = RandomForestClassifier(random_state=42)
clf = LogisticRegression()

clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
y_preds = clf.predict(X_test)
```

```python
print( classification_report(y_test, y_preds) )
```

```python
conf_mat = confusion_matrix(y_test, y_preds)
conf_mat
```
